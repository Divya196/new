Renaming and combining

Renaming the prep_time column to preparation_time.

rename the columns cook_time to cooking_time and diet to diet_type in the DataFrame.

inner join between two DataFrames df1 and df2 on the column region.
(df1:region: ['North', 'South', 'East', 'West'], value1: [10, 20, 30, 40])
(df2:region: ['North', 'East', 'West'], value2: [100, 200, 300])

full outer join between df1 and df2 on the column region.


```python
#Renaming the prep_time column to preparation_time.
indian_food_df.rename(columns ={"prep_time":"preparation_time"}, inplace =True)
indian_food_df
```

```python
#rename the columns cook_time to cooking_time and diet to diet_type in the DataFrame.
indian_food_df.rename(columns={"cook_time":"cooking_time","diet":"diet_type"},inplace = True)
indian_food_df
```

```python
#inner join between two DataFrames df1 and df2 on the column region.
#(df1:region: ['North', 'South', 'East', 'West'], value1: [10, 20, 30, 40])
#(df2:region: ['North', 'East', 'West'], value2: [100, 200, 300])
data_1 = { 'region': ['North', 'South', 'East', 'West'] ,
           'value1': [10, 20, 30, 40]}
df1 = pd.DataFrame(data_1)
df1

data_2 = { 'region': ['North', 'East', 'West'] ,
           'value2' : [100, 200, 300]}
df2 = pd.DataFrame(data_2)
df2

inner_df = pd.merge(df1,df2,how='inner',on = 'region')
inner_df
```

```python
#full outer join between df1 and df2 on the column region.
outer_df = pd.merge(df1,df2,how='outer',on = 'region')
outer_df
```
